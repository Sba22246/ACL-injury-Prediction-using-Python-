```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Load in Data

```python
merged_df=pd.read_csv('merged_df.csv') # containing FM17
```

```python
#Adding colour(String) for visualisation in pywedge later
# Add a new column 'colour' based on conditions in the 'injury_type' column
merged_df['colour'] = np.where(merged_df['injury_type'] == 1, 'blue', 
                               np.where(merged_df['injury_type'] == 0, 'orange', np.nan))
```

```python
#Adding ACL or Non-ACL (String) for pywedge later
# Add a new column 'injury' based on conditions in the 'injury_type' column
merged_df['injury'] = np.where(merged_df['injury_type'] == 1, 'ACL', 
                               np.where(merged_df['injury_type'] == 0, 'Non-ACL', np.nan))
```

```python
merged_df['colour'] = merged_df['colour'].astype(str)
```

```python
# List of columns to drop #Already plotted , will speed up visualisation
cols_to_drop = ['injury_type', 'foot_left', 'foot_right', 'foot_both', 'role_Goalkeeper', 
                'role_Defender', 'role_Midfielder', 'role_Forward']

# Drop the columns from both Dfs
type_1_df = injury_type_1_df.drop(columns=cols_to_drop)
type_0_df = injury_type_0_df.drop(columns=cols_to_drop)

# Plot histograms for Injury Type = 1
type_1_df.hist(figsize=(15, 15), color='red', bins=50, grid=False)
plt.tight_layout()  # Adjusts the subplots to fit
plt.suptitle('Histograms for ACL Injuries', y=1.02)
plt.show()

# Plot histograms for Injury Type = 0
type_0_df.hist(figsize=(15, 15), color='blue', bins=50, grid=False)
plt.tight_layout()  # Adjusts the subplots to fit
plt.suptitle('Histograms for Non-ACL', y=1.02)
plt.show()
```

```python
# ensure common columns
columns = type_1_df.columns

# Determine the number of rows and columns for subplots based on the number of features
n_cols = 3
n_rows = len(columns) // n_cols + (1 if len(columns) % n_cols else 0)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 15))

# Flatten axes for easy iteration
flat_axes = axes.flatten()

for i, column in enumerate(columns):
    ax = flat_axes[i]
    
    # Plot histograms for both injury types
    type_0_df[column].hist(ax=ax, bins=50, color='skyblue', alpha=0.5, label='Injury Type = 0')
    type_1_df[column].hist(ax=ax, bins=50, color='red', alpha=0.5, label='Injury Type = 1')
    
    ax.set_title(column)
    ax.legend()

# Remove any unused subplots
for i in range(len(columns), n_rows * n_cols):
    fig.delaxes(flat_axes[i])

plt.tight_layout()
plt.suptitle('Overlaid Histograms for Both Injury Types', y=1.02)
plt.show()
```

```python

# Loop through each column in the DataFrame
for column in injury_type_0_df.columns:

    # Get value counts for the current column
    feature_counts_0 = injury_type_0_df[column].value_counts().sort_index()
    feature_counts_1 = injury_type_1_df[column].value_counts().sort_index()

    # Harmonizing indices
    unified_index = feature_counts_0.index.union(feature_counts_1.index)
    feature_counts_0 = feature_counts_0.reindex(unified_index, fill_value=0)
    feature_counts_1 = feature_counts_1.reindex(unified_index, fill_value=0)

    # Check if the number of unique values is too high
    # If it's more than 50, skip that column
    if len(unified_index) > 50:
        print(f"Skipped {column} due to too many unique values.")
        continue

    # Calculate bar positions
    bar_width = 0.35
    index = np.arange(len(unified_index))
    
    # Plotting the bar charts for both types on the same graph
    plt.figure(figsize=(12, 8))
    bar1 = plt.bar(index, feature_counts_0, bar_width, color='b', label='Injury Type 0', edgecolor='black')
    bar2 = plt.bar(index + bar_width, feature_counts_1, bar_width, color='r', label='Injury Type 1', edgecolor='black')

    plt.title(f'Counts for {column} by Injury Type')
    plt.xlabel(column)
    plt.ylabel('Counts')
    plt.xticks(index + bar_width / 2, unified_index, rotation=45)  # Set ticks in the middle of grouped bars
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# PYWEDGE
The Pywedge_Charts class takes following inputs,

Inputs:

Dataframe
c = any redundant column to be removed (like ID column etc., at present supports a single column removal, subsequent version will provision multiple column removal requirements)
y = target column name as a string

```python
import plotly.express as px 
import pywedge as pw
```

```python
# Import required libraries


# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(y='BMI', data=merged_df, color='lightblue')

# Calculate statistics for the entire year
Q1 = merged_df['BMI'].quantile(0.25)
Q3 = merged_df['BMI'].quantile(0.75)
median = merged_df['BMI'].median()
min_val = merged_df['BMI'].min()
max_val = merged_df['BMI'].max()

# Add the annotations
plt.annotate('1st Quartile (Q1)', xy=(0, Q1), xytext=(0.5, Q1),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('3rd Quartile (Q3)', xy=(0, Q3), xytext=(0.5, Q3),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('Median', xy=(0, median), xytext=(0.5, median), 
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('Min', xy=(0, min_val), xytext=(0.5, min_val),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('Max', xy=(0, max_val), xytext=(0.5, max_val),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Adjusted arrowprops for the "Width = Frequency" annotation
plt.annotate('Width = Frequency', 
             xy=(0.5, min_val + (max_val-min_val)), 
             xytext=(-0.5, min_val + (max_val-min_val)),  
             )

plt.title('Distribution of BMI for all injuries')
plt.ylabel('BMI')
plt.tight_layout()
plt.show()

```
